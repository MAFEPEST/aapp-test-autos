# Spanish Pipeline (Small)

```python
#Importar Librerías
import spacy
import numpy as np
```

```python
#Cargar Data
data = [
        "Mi mamá ama las manzanas", 
        "Katy Perry odia las manzanas.",
        "Orlando Bloom compró acciones en NotCo",
        "Mi hermana nació en Melipilla",
        "Katy Perry visitó Melipilla",
        "Los fundadores de NotCo nacieron en Melipilla",
        "No me gustó la comida en Melipilla",
        "Orlando Bloom compró una casa en Melipilla",
        "Mi postulación a NotCo fue rechazada",
        "Mi papá fue al colegio en Melipilla",
        "El que fue a Melipilla perdió su silla",
        "No me gustan los productos de NotCo",
        "Anoche hubieron protestas contra NotCo"
        ]
```

```python
#Descargar Pipeline (Spanish)
!python -m spacy download es_core_news_sm -q
```

```python
#Cargar Pipeline
nlp_sm = spacy.load("es_core_news_sm")
```

```python
#Crear Doc Object
doc_sm = nlp_sm(data[0])
```

```python
#Validar Existencia Word Embeddings (GloVe)
for token in doc_sm:
  print(token.text,"---", token.has_vector)
```

```python
#Revisar Representación Vectorial (Token)
print("Dimensionalidad Word Embeddings:", doc_sm[0].vector.shape)
doc_sm[0].vector
```

```python
#Revisar Representación Vectorial (Documento - Promedio Simple)
print("Dimensionalidad Word Embeddings:", doc_sm.vector.shape)
doc_sm.vector
```

```python
#Crear Representación Vectorial para Set Documentos
X_sm = list() #Para cada documento sacar representacion vectorial y guardarlo en una lista
for text in data:
  doc = nlp_sm(text)
  X_sm.append(doc.vector)
```

```python
#Convertir Lista en Numpy Array. Mas facil procesar arreglos en numpy
X_sm_embeddings = np.array(X_sm)
print("Dimensionalidad Embeddings:", X_sm_embeddings.shape)
```
